```python
DIR = './data/10m'
fna = [name for name in os.listdir(DIR)]
fna.sort()
fno = len(fna)
tower = ['East', 'Flux', 'North', 'West']

SweepMean = []
EjectionMean = []
OutwardMean = []
InwardMean = []
OutwardFra = []
EjectionFra = []
InwardFra = []
SweepFra = []
HeatMean = []

for fi in range(5):
    if fi-3:
        if fi<3:
            fi = fi
        elif fi>3:
            fi = fi-1
        xls = pd.ExcelFile(f'./data/10m/{fna[fi]}')
        df_pre = pd.read_excel(xls, 'Pre-FFP')
        df_dur = pd.read_excel(xls, 'FFP')
        df_post = pd.read_excel(xls, 'Post-FFP')

        df = [df_pre, df_dur, df_post]
        for di in range(3):
            SweepMean.append(df[di]["Momentum Ejection (S'<0, w'>0)"][11])
            EjectionMean.append(df[di]["Momentum Ejection (S'<0, w'>0)"][13])
            OutwardMean.append(df[di]["Momentum Ejection (S'<0, w'>0)"][15])
            InwardMean.append(df[di]["Momentum Ejection (S'<0, w'>0)"][17])

            OutwardFra.append(df[di]["Momentum Outward Interaction (S'>0, w'>0)"][7])
            EjectionFra.append(df[di]["Momentum Ejection (S'<0, w'>0)"][7])
            InwardFra.append(df[di]["Momentum Inward Interaction (S'<0, w'<0)"][7])
            SweepFra.append(df[di]["Momentum Sweep (S'>0, w'<0)"][7])
            HeatMean.append(df[di]["Momentum Ejection (S'<0, w'>0)"][19])
    else:
        for di in range(3):
            SweepMean.append(np.nan)
            EjectionMean.append(np.nan)
            OutwardMean.append(np.nan)
            InwardMean.append(np.nan)

            OutwardFra.append(np.nan)
            EjectionFra.append(np.nan)
            InwardFra.append(np.nan)
            SweepFra.append(np.nan)
            HeatMean.append(np.nan)
```

```python
Mean = np.load('./var/Mean.npy')
Freq = np.load('./var/Freq.npy')
```

```python
tower = ['East', 'Flux', 'North', 'South', 'West']
height = ['3m', '10m', '20m']

color = ['lime', 'r', 'k']
event = ['S', 'E', 'I', 'O']
xl, xr, yl, yr = 0, 1, -5, 5
fig, axs = plt.subplots(5, 3, figsize=(15, 20))
for ti in range(5):
    for hi in range(3):
        ax = plt.subplot(5,3,ti*3+hi+1)
        for pi in range(3):
            for ei in range(4):
                ax.arrow(0, 0, Freq[hi, ti, pi, ei], Mean[hi, ti, pi, ei], alpha=0.5, width=0.0001, length_includes_head=True, head_width=0.04, head_length=0.1, color=color[pi])
                ax.text(Freq[hi, ti, pi, ei], Mean[hi, ti, pi, ei], event[ei], fontweight='bold', color=color[pi])
                ax.set_ylim(yl, yr)
                ax.set_xlim(xl, xr)
        plt.hlines(0, xl, xr, 'c', lw=1)
        if ti == 0:
            plt.text(.5, 1.08, height[hi], fontsize='large', fontweight='bold', ha='center', va='center', transform=ax.transAxes)
        if hi == 2:
            plt.text(1.08, .9, tower[ti], fontsize='large', fontweight='bold', ha='center', va='center', transform=ax.transAxes)
        plt.xlabel('Fraction of Event Counts', fontweight='bold')
        plt.ylabel('Contribution of Events', fontweight='bold')
plt.savefig('./plot/ArrowFigure.png', bbox_inches='tight')
```
