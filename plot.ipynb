```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc("savefig", dpi=300)
```

```python
contri = np.load('./data/ContributionValue_Flux.npy')
fracti = np.load('./data/FractionValue_Flux.npy')
```

```python
tower = ['E', 'F', 'N', 'W', 'S']
height = ['3m', '10m', '19m']

color = ['lime', 'r', 'k']
event = ['S', 'E', 'I', 'O']
xl, xr, yl, yr = 0, .7, -1, 5
fig, axs = plt.subplots(1, 3, figsize=(15, 10))
for hi in range(3):
    ax = plt.subplot(1,3,hi+1)
    for ti in range(5):
        for ei in range(4):
            for pi in range(3):
                ax.scatter(fracti[ti, ei, hi*3+pi], contri[ti, ei, hi*3+pi], s=5, c=color[pi])
                ax.text(fracti[ti, ei, hi*3+pi], contri[ti, ei, hi*3+pi], event[ei]+tower[ti], fontsize=7, fontweight='bold', color=color[pi])
    ax.set_ylim(yl, yr)
    ax.set_xlim(xl, xr)
    plt.grid()
    plt.title(height[hi], fontsize=13, fontweight='bold')
    if hi == 0:
        plt.xlabel('Fraction value', fontsize=13, fontweight='bold')
        plt.ylabel('Contribution value', fontsize=13, fontweight='bold')
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, hspace = 0, wspace = 0)
plt.savefig('./plot/paper/ScatterPlot.png',bbox_inches='tight')
```
